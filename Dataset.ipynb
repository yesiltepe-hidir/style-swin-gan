```python
import torch
import torchvision
from torchvision import transforms
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
def get_data_loader(datasetname, root, batch_size):
  if datasetname == 'LSUN':
      dataset = torchvision.datasets.LSUN(
                    root = root,
                    classes = ['church_outdoor_train'],
                    transform = transforms.Compose([
                              transforms.Resize(64),
                              transforms.CenterCrop(64),
                              transforms.ToTensor()
                              ])
                    )
      dataloader = torch.utils.data.DataLoader(
                              dataset,
                              batch_size = batch_size,
                              num_workers = 2,
                              pin_memory = True
      )

  else:
    raise ValueError(f'No dataset named {datasetname}!')
  
  return dataloader
```

```python
# Dataset
datasetname = 'LSUN'
# Data Root
root = 'data/'
# Parameters
batch_size = 256
# Get Dataloader
loader = get_data_loader(datasetname, root, batch_size)
```

```python
def visualize_data(loader, n):
  m = n * n
  for i in range(1, m + 1):
    plt.subplot(n, n, i)
    data = next(iter(loader))[0][i].permute(1, 2, 0)
    print(data.shape)
    print(data.min())
    print(data.max())
    plt.imshow(data)
```

```python
visualize_data(loader, 2)
```

```python
visualize_data(loader, 1)
```
